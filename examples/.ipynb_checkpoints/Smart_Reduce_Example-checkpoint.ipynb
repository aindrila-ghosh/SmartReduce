**Dimensionality reduction** is a commonly used technique in big data analytics. Reducing the dimensionality of datasets not only helps with managing their analytical complexity but also with removing redundancy within them. Over the years, several dimensionality reduction algorithms have been proposed that approach the problem differently. With the same overall goal of preserving the original data structure, these algorithms attempt a linear or a nonlinear transformation on the data, based on the local or global spatial relationships in the dataset. Given a plethora of dimensionality reduction algorithms, there is a long existing need for guidelines on; **How to choose the most appropriate algorithm in a given scenario?** In order to bridge this gap, in this research we perform a large scale experimental study with 15 dimensionally reduction algorithms and compare their performances across 13 contextual metrics using 30 real world datasets. Furthermore, we perform a statistical significance analysis of our obtained results to assess the generalizability of the performances of these algorithms.

For statistical tests: we have **Paired t-test, Wilcoxon's Signed rank test, McNemar's tests, and Friedman's test**. For Post-hoc tests we have **Nemenyi Correction, Holm's Test, Shaffer's Test**.

### Step 4: (Optional) Run Statistical Significance Tests

Once we obtained the values for all the 13 defined metrics for each algorithm on each dataset, we perform statistical significance analysis of the experimental results. Null hypothesis significance testing or statistical significance analysis forms the core of inferential statistics. It is a form of 'reductio ad absurdum' that tries to discredit an idea by assuming the idea is true and then showing a contradiction to the idea. For our experiments, the primary goal of the statistical significance testing is to determine whether there is enough empirical evidence to claim that this difference in the performances of the algorithms is random or statistically significant. 

To enable statistical analysis, it is important to run all the 15 algorithms a large enough number of times on independent samples of all the 30 datasets for all the 7 metrics, so that the probability distribution of the scores for each algorithm on each metric can be understood. In order to perform the null hypothesis significance testing, a null hypothesis (denoted by $H_0$) is defined that states that there is no statistically significant difference between the performances of two algorithms. The alternative hypothesis (represented by $H_a$) on the other hand, states the exact opposite of the null hypothesis, that is, there is a statistically significant difference between the performances of the two algorithms for the selected dataset samples, and hence we must reject $H_0$. 

In order to make a decision on whether or not to reject the null hypothesis for a statistical test, the p-value is computed along with the test statistics. The p-value is the probability of obtaining a result equally or even more extreme, considering $H_0$ is true. In case the p-value is smaller than a predefined nominal significance level $\alpha$, $H_0$ is rejected. A typical value for $\alpha$ that is commonly used in statistical experiments is 0.05. When performing statistical significance testing, one needs to consider two types of possible errors that could occur during the tests. The Type I error represents the situation when an $H_0$ is rejected despite of being true, and the Type II error relates to the situation when $H_0$ is accepted despite of being false. Here, an important point to note is, the two errors contradict each other, as reducing the probability of one error would increase the probability of the other. However, research shows that it is more important to prevent Type I errors than Type II.

The most common way to perform a pairwise comparison of two algorithms is to use the paired t-test. Being a parametric statistical test, the t-test forms the basis of pairwise comparisons among algorithms, that examines whether the mean difference between the performances of the two algorithms is significantly greater than 0. Formally, the t-statistic is computed as:

$$ t = \frac{\bar{d_n}_a}{{{\sigma_d}_n}_a} $$

Where, $\bar{d_n}_a$ represents the mean difference of all $n_a$ datasets,and ${{\sigma_d}_n}_a$ represents their standard deviation. The t-statistic follows the Studentâ€™s t-distribution with N-1 degrees of freedom for N datasets and in case the p-value corresponding to the t-statistic is less than the threshold $\alpha$, the null hypothesis  is rejected.

In order to avoid the normality assumptions of the paired t-test, research shows that the Wilcoxon signed ranks test is its most appropriate non-parametric counterpart. The Wilcoxon signed ranks test is used to compare the absolute differences in performances of the two algorithms to determine if the mean ranks of the positive and negative differences between the performance scores are statistically significant. Formally, ranking the differences based on their absolute values, and assuming $R^+$ be the sum of ranks where ${d_n}_a > 0$ and $R^-$ be the sum of ranks where ${d_n}_a < 0$; We can define ranks $R^+$ and $R^-$ as:

$$ R^+ = \sum_{{d_n}_a > 0} rank({d_n}_a) + \frac{1}{2} \sum_{{d_n}_a = 0} rank({d_n}_a) $$
$$ R^- = \sum_{{d_n}_a < 0} rank({d_n}_a) + \frac{1}{2} \sum_{{d_n}_a = 0} rank({d_n}_a) $$
		 		 
Considering $T = min($R^+$,$R^-$)$, the test statistic z for the Wilcoxon signed ranks test for number of datasets N > 25 is calculated as:

$$ z = \frac{T - \frac{1}{4} N(N+1)}{\sqrt{\frac{1}{24}N(N+1)(2N+1)}} $$

The test statistic for Wilcoxon signed ranks test approximately follows the normal distribution. The null hypothesis $H_0$ is rejected in case the p-value corresponding to the test statistic is less than the threshold $\alpha$. 

The Wilcoxon signed ranks test is safer than paired t-tests as not only they do not make any assumptions regarding the normal distribution of the data, but also, outliers (i.e., an unusually good or bad performances on any particular dataset) do not impact the Wilcoxon test as much as the paired t-test.

In order to validate the efficiency of the Wilcoxon test, we also calculate the McNemarâ€™s exact and asymptotic test statistics for pairwise comparisons among the algorithms. When comparing two algorithms for a metric over N datasets, considering $n_01$ is the number of datasets where the performance of the first algorithm is better than the second and $n_10$ is otherwise, The McNemarâ€™s asymptotic test statistic is calculated as:

$$ \chi^2 = \frac{(|n_01 - n_10|)^2}{n_01+n_10} $$
						
The statistic follows a  distribution with one degree of freedom under the assumption that the null hypothesis is true. Nevertheless, arguably the most popular version of the McNemarâ€™s test statistic is the McNemarâ€™s exact test, that can be measured as follows:

$$ \chi^2 = \frac{(|n_01 - n_10|-1)^2}{n_01+n_10} $$

Although, the McNemarâ€™s exact test is known for its conservative nature and the resultant type I and type II errors, we used the statistic in our calculations firstly because of its popularity and secondly to validate our experimental outcome with conservative tests as well.

When comparing multiple algorithms, in our experiments we used the Friedman's statistics. In the Friedman test, instead of using the actual performance scores of the algorithms, for evaluating their performances, at first the scores are ranked based on their value with 1 being the highest rank. For the algorithms that produce the same scores, average ranks are assigned to them. Formally, the Friedman statistic compares the mean ranks of all algorithms on a metric. The Friedman statistic was defined by Iman and Davenport as:

$$ F_F = \frac{(N-1)\chi^2F}{N(K-1)-\chi^2F} $$

However, in this situation due to the challenge of multiple hypotheses testing, adjustments to the p-values and the threshold $\alpha$ are made. The most common adjustment to the p-values is the Nemenyi correction. In this adjustment, the threshold  and the p-values are divided by the total number of pairwise comparisons among the algorithms. Nevertheless, Nemenyi adjustment being conservative, in some cases it can increase the probability of the Type II error. Hence, an alternative to the Nemenyi correction is the Holm procedure. This procedure iteratively selects the most significant p-value among all the test statistics of pairwise combinations, and divides the p-value with $\frac{\alpha}{\kappa - 1}$. The corresponding null hypothesis is rejected in case the selected p-value is less than $\frac{\alpha}{\kappa - 1}$. In the next iteration, the next most significant p-value is selected. In our experiments for the post-hoc tests for the Friedman test we perform both Nemenyi and Holm adjustments and compare the results.
